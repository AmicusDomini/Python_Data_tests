```python
import pandas as pd

ay = pd.read_csv('hf://datasets/lukebarousse/data_jobs/data_jobs.csv')

ay['job_posted_date'] = pd.to_datetime (ay['job_posted_date'])
```

```python
ay.groupby('job_title_short')['salary_year_avg'].median()
```

```python
top_countries = ay['job_country'].value_counts().head(6).index

ay_job_country_salary = ay.pivot_table (
    values='salary_year_avg',
    index= 'job_country',
    columns='job_title_short',
    aggfunc='median'
)

ay_job_country_salary = ay_job_country_salary.loc[top_countries]

ay_job_country_salary
```

```python
job_titles = ['Senior Data Analyst', 'Senior Data Engineer', 'Senior Data Scientist']

ay_job_country_salary = ay_job_country_salary[job_titles]

ay_job_country_salary
```

```python
ay_job_country_salary.plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

ay_job_country_salary.plot(kind='bar')
plt.ylabel ('Median Salary ($)')
plt.xlabel ('')
plt.title ('Median Salary by Country and Job Title')
plt.xticks (rotation = 45, ha = 'right')
plt.show ()
```
