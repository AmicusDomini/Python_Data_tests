```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

ay = pd.read_csv('hf://datasets/lukebarousse/data_jobs/data_jobs.csv')

ay['job_posted_date'] = pd.to_datetime (ay['job_posted_date'])

ay['job_skills']= ay['job_skills'].apply(lambda skill_list: 
                                         ast.literal_eval(skill_list) 
                                         if pd.notna (skill_list) else skill_list)
```

```python
ay_DA= ay[ay['job_title_short']=='Data Analyst'].copy()

ay_DA
```

```python
ay_DA['job_posted_month']=ay_DA['job_posted_date'].dt.month

ay_DA.info()
```

```python
ay_DA_explode= ay_DA.explode('job_skills')
```

```python
ay_DA_explode
```

```python
ay_DA_expivot = ay_DA_explode.pivot_table(index='job_posted_month', columns='job_skills',aggfunc= 'size', fill_value=0)
ay_DA_expivot
```

```python
ay_DA_expivot.loc['Total']=ay_DA_expivot.sum()

ay_DA_expivot
```

```python
ay_DA_expivot=ay_DA_expivot[ay_DA_expivot.loc['Total'].sort_values(ascending=False).index]

ay_DA_expivot= ay_DA_expivot.drop('Total')

ay_DA_expivot
```

```python
ay_DA_expivot.iloc[:,0:7].plot(kind= 'line')

plt.ylabel('Count')
plt.title('Top 7 skills for Data Analysts in 2023')
plt.xlabel('')
plt.show()
```

```python
ay_DA_expivot = ay_DA_expivot.reset_index(names='month name')
ay_DA_expivot['month name']=ay_DA_expivot['month name'].apply(
    lambda x: pd.to_datetime(x, format= '%m').strftime ('%b')
    )
```

```python
ay_DA_expivot = ay_DA_expivot.set_index('month name') 
#ie to remove the former index and make 'month name' column as new index ie the first column. Then,
ay_DA_expivot.iloc[:,0:7].plot(kind= 'line')

plt.ylabel('Count')
plt.title('Top 7 skills for Data Analysts in 2023')
plt.xlabel('')
plt.show()
```
