```python
import pandas as pd
import numpy as np

# Read the first 'n' rows of a CSV file
n = 500000  # Specify the number of rows you want
df = pd.read_csv('hf://datasets/lukebarousse/data_jobs/data_jobs.csv', nrows=n)

# Display the first 'n' rows
(df)
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# df.iloc[2] #displays the 2nd row
df.iloc [0:10] #displays first 10 rows (row 0 to 10)
```

```python
df.loc [0:10, "salary_rate":"salary_hour_avg"]
```

```python
df.loc [:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

```python
median_yrsalary_avg = df['salary_year_avg'].median ()
median_hrsalary_avg = df['salary_hour_avg'].median ()
```

```python
import numpy as np
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_yrsalary_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hrsalary_avg)
```

```python
df_filled.loc [:10, 'salary_rate':'salary_hour_avg']
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()
```

```python
print (len(df_filled))
print (len (df_unique))
print ('Rows dropped  =', len (df_filled)- len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

print (len(df_filled))
print (len (df_unique))
print ('Rows dropped  =', len (df_filled)- len(df_unique))
```
